```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
```

```python
!ls
```

```python
first_half = sqlite3.connect("../../databases/last_attempt.db")
second_half = sqlite3.connect("../../databases/last_attempt.db")
```

```python
first = pd.read_sql_query("select * from sneakers;", first_half)
second = pd.read_sql_query("select * from sneakers;", second_half)
other = pd.read_csv("../data/sneakers.csv")
```

```python
first
```

```python
first = first.drop_duplicates(subset="sneaker", keep="first")
second = second.drop_duplicates(subset="sneaker", keep="first")
other = other.drop_duplicates(subset="sneaker", keep="first")
```

```python
price = first.iloc[:,2]
brands = first.iloc[:,3]
years = first.iloc[:,4].str[17:-2]
```

```python
shoe_data_by_year = list(zip(brands, price, years))
first_df = pd.DataFrame(shoe_data_by_year, columns = ['brand', 'price', 'year'])
```

```python
first_df
```

```python
second
```

```python
second_price = second.iloc[:,2]
second_brands = second.iloc[:,3]
second_years = second.iloc[:,4].str[17:-2]
```

```python
sec_shoe_data_by_year = list(zip(second_brands, second_price, second_years))
second_df = pd.DataFrame(sec_shoe_data_by_year, columns = ['brand', 'price', 'year'])
```

```python
second_df
```

```python
other = other.drop("Unnamed: 0", axis=1)
other
```

```python
other_price = other.iloc[:,2]
other_brands = other.iloc[:,3]
other_years = other.iloc[:,4].str[15:]
```

```python
other_shoe_data_by_year = list(zip(other_brands, other_price, other_years))
other_df = pd.DataFrame(other_shoe_data_by_year, columns = ['brand', 'price', 'year'])
```

```python
other_df
```

```python
sneakers = pd.concat([first_df, second_df, other_df])
```

```python
sneakers
```

```python
sneakers = sneakers.astype({'price': 'int32'})
```

```python
sneakers = sneakers.astype({'brand': 'str'})
```

```python
sneakers = sneakers.astype({'year': 'int32'})
```

```python
sneakers.info()
```

```python
sneakers.to_csv("year_data.csv")
```
